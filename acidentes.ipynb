```python
import pandas as pd
import numpy as np

df= pd.read_csv("US_Accidents_1_milhao.csv")
```

# 1. Carregar o dataset

```python
print("Tamanho original da base:")
print(df.shape)
```

```python
print("\nPrimeiras linhas:")
df.head()
```

```python
print("\nColunas:")
print(df.columns.tolist())
```

# 2. Remover colunas desnecessárias

```python
colunas_remover = [
    "ID",
    "Source",
    "Description",
    "Street",
    "Zipcode",
    "Airport_Code",
    "Country",
    "End_Lat",
    "End_Lng",
    "Weather_Timestamp",
    "Timezone"
]

df = df.drop(columns=colunas_remover, errors="ignore")

```

```python
print(df.shape)
```

# 3. Verificar valores ausentes

```python
valores_ausentes = df.isnull().sum()
percentual_ausentes = (df.isnull().mean() * 100).round(2)
```

```python
tabela_ausentes = pd.DataFrame({
    "valores_ausentes": valores_ausentes,
    "percentual": percentual_ausentes
})
```

```python
tabela_ausentes = tabela_ausentes[tabela_ausentes["valores_ausentes"] > 0]
tabela_ausentes = tabela_ausentes.sort_values(by="percentual", ascending=False)
```

```python
print("\nValores ausentes:")
print(tabela_ausentes)
```

```python
print("Ausentes em Precipitation(in):")
print(df["Precipitation(in)"].isnull().mean() * 100)

df["Precipitation_Ausente"] = df["Precipitation(in)"].isnull()

clima_precipitacao_ausente = (
    df[df["Precipitation_Ausente"] == True]["Weather_Condition"]
    .value_counts(normalize=True)
    .head(20) * 100
)

print("\nCondições climáticas quando Precipitation está ausente:")
print(clima_precipitacao_ausente)
```

```python
df["Precipitation_Ausente"] = df["Precipitation(in)"].isnull()
df["Precipitation(in)"] = df["Precipitation(in)"].fillna(0)
```

```python
print("Ausentes em Wind_Chill(F):")
print(df["Wind_Chill(F)"].isnull().mean() * 100)

print("\nTemperatura média quando Wind_Chill está ausente:")
print(df[df["Wind_Chill(F)"].isnull()]["Temperature(F)"].mean())

print("\nTemperatura média quando Wind_Chill está presente:")
print(df[df["Wind_Chill(F)"].notnull()]["Temperature(F)"].mean())

print("\nVelocidade média do vento quando Wind_Chill está ausente:")
print(df[df["Wind_Chill(F)"].isnull()]["Wind_Speed(mph)"].mean())

print("\nVelocidade média do vento quando Wind_Chill está presente:")
print(df[df["Wind_Chill(F)"].notnull()]["Wind_Speed(mph)"].mean())
```

```python
df["Wind_Chill_Ausente"] = df["Wind_Chill(F)"].isnull()

mediana_temperatura = df["Temperature(F)"].median()

df["Temperature(F)"] = df["Temperature(F)"].fillna(mediana_temperatura)

df["Wind_Chill(F)"] = df["Wind_Chill(F)"].fillna(df["Temperature(F)"])
```

```python
print(df["Temperature(F)"].isnull().sum())
print(df["Wind_Chill(F)"].isnull().sum())
print(df["Wind_Chill_Ausente"].value_counts())
```

```python
print("Ausentes em Wind_Speed(mph):")
print(df["Wind_Speed(mph)"].isnull().mean() * 100)

print("\nDireção do vento quando Wind_Speed está ausente:")
print(
    df[df["Wind_Speed(mph)"].isnull()]["Wind_Direction"]
    .value_counts(normalize=True)
    .head(15) * 100
)

print("\nCondição climática quando Wind_Speed está ausente:")
print(
    df[df["Wind_Speed(mph)"].isnull()]["Weather_Condition"]
    .value_counts(normalize=True)
    .head(15) * 100
)
```

```python
df["Wind_Speed_Ausente"] = df["Wind_Speed(mph)"].isnull()

df.loc[
    (df["Wind_Speed(mph)"].isnull()) & (df["Wind_Direction"] == "Calm"),
    "Wind_Speed(mph)"
] = 0

mediana_vento = df["Wind_Speed(mph)"].median()
df["Wind_Speed(mph)"] = df["Wind_Speed(mph)"].fillna(mediana_vento)
```

```python
colunas_numericas_mediana = [
    "Visibility(mi)",
    "Humidity(%)",
    "Temperature(F)",
    "Pressure(in)"
]

for coluna in colunas_numericas_mediana:
    mediana = df[coluna].median()
    df[coluna] = df[coluna].fillna(mediana)
```

```python
colunas_categoricas_desconhecido = [
    "Weather_Condition",
    "Wind_Direction",
    "Sunrise_Sunset",
    "Nautical_Twilight",
    "Civil_Twilight",
    "Astronomical_Twilight",
    "City"
]

for coluna in colunas_categoricas_desconhecido:
    df[coluna] = df[coluna].fillna("Desconhecido")
```
