```python
import pandas as pd
import numpy as np

df= pd.read_csv("US_Accidents_1_milhao.csv")
```

# 1. Carregar o dataset

```python
print("Tamanho original da base:")
print(df.shape)
```

```python
print("\nPrimeiras linhas:")
df.head()
```

```python
print("\nColunas:")
print(df.columns.tolist())
```

# 2. Remover colunas desnecessárias

```python
colunas_remover = [
    "ID",
    "Source",
    "Description",
    "Street",
    "Zipcode",
    "Airport_Code",
    "Country",
    "End_Lat",
    "End_Lng",
    "Weather_Timestamp",
    "Timezone"
]

df = df.drop(columns=colunas_remover, errors="ignore")

```

```python
print(df.shape)
```

# 3. Verificar valores ausentes

```python
valores_ausentes = df.isnull().sum()
percentual_ausentes = (df.isnull().mean() * 100).round(2)
```

```python
tabela_ausentes = pd.DataFrame({
    "valores_ausentes": valores_ausentes,
    "percentual": percentual_ausentes
})
```

```python
tabela_ausentes = tabela_ausentes[tabela_ausentes["valores_ausentes"] > 0]
tabela_ausentes = tabela_ausentes.sort_values(by="percentual", ascending=False)
```

```python
print("\nValores ausentes:")
print(tabela_ausentes)
```

```python
print("Ausentes em Precipitation(in):")
print(df["Precipitation(in)"].isnull().mean() * 100)

df["Precipitation_Ausente"] = df["Precipitation(in)"].isnull()

clima_precipitacao_ausente = (
    df[df["Precipitation_Ausente"] == True]["Weather_Condition"]
    .value_counts(normalize=True)
    .head(20) * 100
)

print("\nCondições climáticas quando Precipitation está ausente:")
print(clima_precipitacao_ausente)
```

```python
df["Precipitation_Ausente"] = df["Precipitation(in)"].isnull()
df["Precipitation(in)"] = df["Precipitation(in)"].fillna(0)
```

```python
print("Ausentes em Wind_Chill(F):")
print(df["Wind_Chill(F)"].isnull().mean() * 100)

print("\nTemperatura média quando Wind_Chill está ausente:")
print(df[df["Wind_Chill(F)"].isnull()]["Temperature(F)"].mean())

print("\nTemperatura média quando Wind_Chill está presente:")
print(df[df["Wind_Chill(F)"].notnull()]["Temperature(F)"].mean())

print("\nVelocidade média do vento quando Wind_Chill está ausente:")
print(df[df["Wind_Chill(F)"].isnull()]["Wind_Speed(mph)"].mean())

print("\nVelocidade média do vento quando Wind_Chill está presente:")
print(df[df["Wind_Chill(F)"].notnull()]["Wind_Speed(mph)"].mean())
```

```python
df["Wind_Chill_Ausente"] = df["Wind_Chill(F)"].isnull()

mediana_temperatura = df["Temperature(F)"].median()

df["Temperature(F)"] = df["Temperature(F)"].fillna(mediana_temperatura)

df["Wind_Chill(F)"] = df["Wind_Chill(F)"].fillna(df["Temperature(F)"])
```

```python
print(df["Temperature(F)"].isnull().sum())
print(df["Wind_Chill(F)"].isnull().sum())
print(df["Wind_Chill_Ausente"].value_counts())
```

```python
print("Ausentes em Wind_Speed(mph):")
print(df["Wind_Speed(mph)"].isnull().mean() * 100)

print("\nDireção do vento quando Wind_Speed está ausente:")
print(
    df[df["Wind_Speed(mph)"].isnull()]["Wind_Direction"]
    .value_counts(normalize=True)
    .head(15) * 100
)

print("\nCondição climática quando Wind_Speed está ausente:")
print(
    df[df["Wind_Speed(mph)"].isnull()]["Weather_Condition"]
    .value_counts(normalize=True)
    .head(15) * 100
)
```

```python
df["Wind_Speed_Ausente"] = df["Wind_Speed(mph)"].isnull()

df.loc[
    (df["Wind_Speed(mph)"].isnull()) & (df["Wind_Direction"] == "Calm"),
    "Wind_Speed(mph)"
] = 0

mediana_vento = df["Wind_Speed(mph)"].median()
df["Wind_Speed(mph)"] = df["Wind_Speed(mph)"].fillna(mediana_vento)
```

```python
colunas_numericas_mediana = [
    "Visibility(mi)",
    "Humidity(%)",
    "Temperature(F)",
    "Pressure(in)"
]

for coluna in colunas_numericas_mediana:
    mediana = df[coluna].median()
    df[coluna] = df[coluna].fillna(mediana)
```

```python
colunas_categoricas_desconhecido = [
    "Weather_Condition",
    "Wind_Direction",
    "Sunrise_Sunset",
    "Nautical_Twilight",
    "Civil_Twilight",
    "Astronomical_Twilight",
    "City"
]

for coluna in colunas_categoricas_desconhecido:
    df[coluna] = df[coluna].fillna("Desconhecido")
```

# Análise Exploratória de Dados (EDA)

```python
# Análise Exploratória de Dados (EDA)
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("Set2")
```

## Distribuições das variáveis quantitativas

```python
# Lista das variáveis numéricas disponíveis após limpeza
variaveis_numericas = [
    'Distance(mi)', 'Temperature(F)', 'Wind_Chill(F)',
    'Humidity(%)', 'Pressure(in)', 'Visibility(mi)',
    'Wind_Speed(mph)', 'Precipitation(in)'
]

fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

for i, var in enumerate(variaveis_numericas):
    axes[i].hist(df[var], bins=50, edgecolor='black', alpha=0.7, color='steelblue')
    axes[i].set_title(f'Distribuição de {var}', fontsize=11)
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frequência')

    mediana = df[var].median()
    axes[i].axvline(mediana, color='red', linestyle='--', linewidth=1.5, label=f'Mediana: {mediana:.2f}')
    axes[i].legend(fontsize=8)

plt.suptitle('Histogramas das Variáveis Numéricas', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('distribuicoes_numericas.png', dpi=150)
plt.show()
```

## Boxplots para visualizar outliers

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 10))
axes = axes.flatten()

for i, var in enumerate(variaveis_numericas):
    axes[i].boxplot(df[var].dropna(), vert=True, patch_artist=True,
                    boxprops=dict(facecolor='lightblue'))
    axes[i].set_title(f'Boxplot de {var}', fontsize=11)
    axes[i].set_ylabel(var)

plt.suptitle('Boxplots das Variáveis Numéricas', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('boxplots_numericas.png', dpi=150)
plt.show()
```

## Análise da variável alvo (Severity)

```python
contagem_severidade = df['Severity'].value_counts().sort_index()
percentual_severidade = df['Severity'].value_counts(normalize=True).sort_index() * 100

tabela_severidade = pd.DataFrame({
    'Contagem': contagem_severidade,
    'Percentual (%)': percentual_severidade.round(2)
})
print("Distribuição original da severidade:")
print(tabela_severidade)

fig, ax = plt.subplots(figsize=(8, 5))
barras = ax.bar(contagem_severidade.index.astype(str), contagem_severidade.values,
                color='steelblue', edgecolor='black')
ax.set_title('Distribuição da Severidade dos Acidentes')
ax.set_xlabel('Severidade (1 = menos grave, 4 = mais grave)')
ax.set_ylabel('Número de Acidentes')

for barra in barras:
    altura = barra.get_height()
    ax.annotate(f'{altura:,}', xy=(barra.get_x() + barra.get_width()/2, altura),
                xytext=(0, 3), textcoords="offset points", ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('distribuicao_severidade.png', dpi=150)
plt.show()
```

## Criação da variável alvo binária (grave = Severity 3 ou 4)

```python
# grave_alta = 1 para Severity 3 ou 4 (acidente grave)
# grave_alta = 0 para Severity 1 ou 2 (acidente leve)
df['grave_alta'] = (df['Severity'] >= 3).astype(int)

print("Distribuição da variável binária grave_alta:")
print(df['grave_alta'].value_counts())
print(f"\nAcidentes graves (Severity 3-4): {df['grave_alta'].mean() * 100:.2f}%")
print(f"Acidentes leves (Severity 1-2): {(1 - df['grave_alta'].mean()) * 100:.2f}%")

fig, ax = plt.subplots(figsize=(6, 6))
labels = ['Leve (Severity 1-2)', 'Grave (Severity 3-4)']
tamanhos = [(df['grave_alta'] == 0).sum(), (df['grave_alta'] == 1).sum()]
cores = ['#66b3ff', '#ff9999']
ax.pie(tamanhos, labels=labels, colors=cores, autopct='%1.1f%%', startangle=90, explode=(0.05, 0.05))
ax.set_title('Proporção de Acidentes Leves vs Graves')
plt.savefig('proporcao_grave_alta.png', dpi=150)
plt.show()
```

## Relação entre variáveis categóricas e gravidade

```python
variaveis_categoricas = [
    'Sunrise_Sunset', 'Crossing', 'Junction', 'Traffic_Signal',
    'Amenity', 'Stop', 'Railway', 'Roundabout'
]

fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.flatten()

for i, var in enumerate(variaveis_categoricas):
    proporcao_grave = df.groupby(var)['grave_alta'].mean().sort_values(ascending=False)
    axes[i].barh(range(len(proporcao_grave)), proporcao_grave.values, color='coral', edgecolor='black')
    axes[i].set_yticks(range(len(proporcao_grave)))
    axes[i].set_yticklabels(proporcao_grave.index)
    axes[i].set_xlabel('Proporção de Acidentes Graves')
    axes[i].set_title(var)
    axes[i].axvline(df['grave_alta'].mean(), color='blue', linestyle='--', label='Média geral')
    axes[i].legend(fontsize=8)

plt.suptitle('Impacto das Características da Via na Gravidade', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('categoricas_vs_gravidade.png', dpi=150)
plt.show()

# Exibir números
print("\nProporção de acidentes graves por característica:")
for var in variaveis_categoricas:
    print(f"\n=== {var} ===")
    print(df.groupby(var)['grave_alta'].mean().round(3) * 100)
```

## Relação entre condições climáticas e gravidade

```python
# Selecionar as 10 condições climáticas mais frequentes
condicoes_frequentes = df['Weather_Condition'].value_counts().head(10).index
df_clima_freq = df[df['Weather_Condition'].isin(condicoes_frequentes)]
proporcao_grave_clima = df_clima_freq.groupby('Weather_Condition')['grave_alta'].mean().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(12, 6))
barras = ax.barh(range(len(proporcao_grave_clima)), proporcao_grave_clima.values, color='teal', edgecolor='black')
ax.set_yticks(range(len(proporcao_grave_clima)))
ax.set_yticklabels(proporcao_grave_clima.index)
ax.set_xlabel('Proporção de Acidentes Graves')
ax.set_title('Acidentes Graves por Condição Climática')
ax.axvline(df['grave_alta'].mean(), color='red', linestyle='--',
           label=f"Média geral: {df['grave_alta'].mean()*100:.1f}%")
ax.legend()

for i, barra in enumerate(barras):
    largura = barra.get_width()
    ax.annotate(f'{largura*100:.1f}%', xy=(largura + 0.005, barra.get_y() + barra.get_height()/2),
                va='center', fontsize=9)

plt.tight_layout()
plt.savefig('clima_vs_gravidade.png', dpi=150)
plt.show()

print(f"\nCondição com maior proporção de graves: {proporcao_grave_clima.index[0]} ({proporcao_grave_clima.iloc[0]*100:.1f}%)")
print(f"Condição com menor proporção de graves: {proporcao_grave_clima.index[-1]} ({proporcao_grave_clima.iloc[-1]*100:.1f}%)")
```

## Análise temporal (hora do dia e mês)

```python
# Converter Start_Time para datetime se necessário
if not pd.api.types.is_datetime64_any_dtype(df['Start_Time']):
    df['Start_Time'] = pd.to_datetime(df['Start_Time'])

df['hora'] = df['Start_Time'].dt.hour
df['mes'] = df['Start_Time'].dt.month

# Mapear meses para estações em português
estacoes_map = {12: 'Inverno', 1: 'Inverno', 2: 'Inverno',
                3: 'Primavera', 4: 'Primavera', 5: 'Primavera',
                6: 'Verão', 7: 'Verão', 8: 'Verão',
                9: 'Outono', 10: 'Outono', 11: 'Outono'}
df['estacao'] = df['mes'].map(estacoes_map)
```

```python
acidentes_por_hora = df.groupby('hora').size()
graves_por_hora = df.groupby('hora')['grave_alta'].mean()

fig, ax1 = plt.subplots(figsize=(14, 6))

# Barras: total de acidentes
ax1.bar(acidentes_por_hora.index, acidentes_por_hora.values, alpha=0.6,
        color='lightblue', label='Total de acidentes')
ax1.set_xlabel('Hora do Dia')
ax1.set_ylabel('Número de Acidentes', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Linha: proporção de graves
ax2 = ax1.twinx()
ax2.plot(graves_por_hora.index, graves_por_hora.values * 100,
         color='red', marker='o', linewidth=2, label='% Acidentes Graves')
ax2.set_ylabel('Proporção de Acidentes Graves (%)', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.axhline(df['grave_alta'].mean() * 100, color='orange', linestyle='--',
            alpha=0.7, label='Média geral')

fig.suptitle('Acidentes por Hora do Dia: Volume vs Gravidade')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.savefig('acidentes_por_hora.png', dpi=150)
plt.show()

# Insights temporais
madrugada = df['hora'].between(0, 5)
print(f"Proporção de graves na madrugada (0-5h): {df[madrugada]['grave_alta'].mean()*100:.1f}%")
print(f"Proporção de graves no restante do dia: {df[~madrugada]['grave_alta'].mean()*100:.1f}%")
```

```python
acidentes_por_mes = df.groupby('mes').size()
graves_por_mes = df.groupby('mes')['grave_alta'].mean()

fig, ax1 = plt.subplots(figsize=(12, 5))
ax1.bar(acidentes_por_mes.index, acidentes_por_mes.values, alpha=0.6,
        color='lightgreen', label='Total de acidentes')
ax1.set_xlabel('Mês')
ax1.set_ylabel('Número de Acidentes', color='green')
ax1.tick_params(axis='y', labelcolor='green')

ax2 = ax1.twinx()
ax2.plot(graves_por_mes.index, graves_por_mes.values * 100,
         color='purple', marker='s', linewidth=2, label='% Acidentes Graves')
ax2.set_ylabel('Proporção de Acidentes Graves (%)', color='purple')
ax2.tick_params(axis='y', labelcolor='purple')

plt.title('Acidentes por Mês: Volume vs Gravidade')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.tight_layout()
plt.savefig('acidentes_por_mes.png', dpi=150)
plt.show()

print("\nAcidentes por estação:")
print(df.groupby('estacao')['grave_alta'].agg(['count', 'mean']).round(3))
```

## Matriz de correlação

```python
# Selecionar variáveis numéricas para correlação
var_correlacao = [
    'grave_alta', 'Distance(mi)', 'Temperature(F)',
    'Humidity(%)', 'Pressure(in)', 'Visibility(mi)',
    'Wind_Speed(mph)', 'Precipitation(in)'
]

df_corr = df[var_correlacao].copy()
df_corr = df_corr.replace([np.inf, -np.inf], np.nan).dropna()
matriz_corr = df_corr.corr()

fig, ax = plt.subplots(figsize=(10, 8))
mascara = np.triu(np.ones_like(matriz_corr, dtype=bool), k=1)
sns.heatmap(matriz_corr, mask=mascara, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, square=True, linewidths=0.5, ax=ax,
            cbar_kws={'shrink': 0.8, 'label': 'Correlação'})
ax.set_title('Matriz de Correlação entre Variáveis Numéricas')

plt.tight_layout()
plt.savefig('matriz_correlacao.png', dpi=150)
plt.show()

print("\nCorrelação com grave_alta (acidente grave):")
corr_com_alvo = matriz_corr['grave_alta'].sort_values(ascending=False)
for var, valor in corr_com_alvo.items():
    print(f"  {var}: {valor:.3f}")
```

## Análise de infraestrutura viária

```python
var_infra = ['Crossing', 'Junction', 'Traffic_Signal', 'Stop', 'Railway', 'Roundabout']
df['num_infra'] = df[var_infra].sum(axis=1)

print("Impacto da quantidade de infraestrutura na gravidade:")
print(df.groupby('num_infra')['grave_alta'].agg(['count', 'mean']).round(3) * 100)

fig, ax = plt.subplots(figsize=(10, 5))
proporcao_grave_infra = df.groupby('num_infra')['grave_alta'].mean()
ax.bar(proporcao_grave_infra.index, proporcao_grave_infra.values * 100,
       color='darkorange', edgecolor='black')
ax.set_xlabel('Quantidade de Características de Infraestrutura')
ax.set_ylabel('Proporção de Acidentes Graves (%)')
ax.set_title('Infraestrutura Viária vs Gravidade do Acidente')
ax.axhline(df['grave_alta'].mean() * 100, color='red', linestyle='--', label='Média geral')
ax.legend()

for i, v in enumerate(proporcao_grave_infra.values):
    ax.text(i, v * 100 + 0.5, f'{v*100:.1f}%', ha='center', fontsize=10)

plt.tight_layout()
plt.savefig('infraestrutura_vs_gravidade.png', dpi=150)
plt.show()

print(f"\nLocais com cruzamento (Crossing=True) têm {df[df['Crossing']==True]['grave_alta'].mean()*100:.1f}% de graves")
print(f"Locais sem cruzamento têm {df[df['Crossing']==False]['grave_alta'].mean()*100:.1f}% de graves")
```
